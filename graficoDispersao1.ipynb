```
import pandas as pd
import csv
import sklearn
import matplotlib.pyplot as plt
import numpy as np
```

```
df = pd.read_csv ('/content/drive/My Drive/PROJETO/2020/dataGraf.csv')
```

```
df.head()
```

##Gráfico dos Negativos


```
import matplotlib.ticker as ticker
```

```
fig, ax = plt.subplots(figsize=(30, 10))
ax.plot(df.data, df.negativo, 'red', linestyle='-')

fig.suptitle("Gráfico de Dispersão - Negativos", fontsize=30)

ax.set_ylabel('Quantidade', fontsize=20)
ax.set_xlabel('Data', fontsize=20)

ax.tick_params(labelsize=15)

tick_spacing = 7
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

plt.show()
```

##Gráfico dos Positivos

```
fig, ax = plt.subplots(figsize=(30, 10))
ax.plot(df.data, df.positivo, 'blue', linestyle='-')

fig.suptitle("Gráfico de Dispersão - Positivos", fontsize=30)

ax.set_ylabel('Quantidade', fontsize=20)
ax.set_xlabel('Data', fontsize=20)

ax.tick_params(labelsize=15)

tick_spacing = 7
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

plt.show()
```

##Gráfico Negativos e Positivos

```
fig, ax = plt.subplots(figsize=(30, 10))
ax.plot(df.data, df.positivo, 'blue', linestyle='--', label='Positivo')
ax.plot(df.data, df.negativo, 'red', linestyle='-', label='Negativo')

ax.legend(loc='upper right', prop={'size': 15})

fig.suptitle("Gráfico de Dispersão - Positivos", fontsize=30)

ax.set_ylabel('Quantidade', fontsize=20)
ax.set_xlabel('Data', fontsize=20)

ax.tick_params(labelsize=15)

tick_spacing = 7
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

plt.show()
```
